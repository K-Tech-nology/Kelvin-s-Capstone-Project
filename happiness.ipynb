```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

df = pd.read_csv('world-happiness-report-2021.csv')
df.head()
```

```python
df.info()
```

```python
pd.set_option('display.max_rows',None,'display.max_columns',None)
df.sort_values(by = ['Ladder score'], ascending = True)
```

```python
import seaborn as sns
corr = df.corr()
ax = sns.heatmap(
    corr,
    vmin = -1, vmax = 1, center = 0,
    cmap = sns.diverging_palette(20,220,n=200),
    square = True
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 90,
    horizontalalignment = 'right'
)
```

```python
#Factors that have high correlation w/ happiness

factor = ['Logged GDP per capita','Social support','Healthy life expectancy']
result = []
for a in factor :
    result.append(df.groupby(['Regional indicator'], as_index=False)[a].mean().sort_values(by = a,ascending = False))
    
print(result)
```

```python
gdp = result[0]
ss = result[1]
le = result[2]

df_ = [gdp,ss,le]

for a in df_ :
    a.columns = a.columns.str.title()
  
#Visualize Result
fig, axes = plt.subplots(1,3)

axes[0].bar(gdp.iloc[:,0],gdp.iloc[:,1], color = 'green')
axes[0].set_xlabel(gdp.columns[0])
axes[0].set_ylabel(gdp.columns[1])

axes[1].bar(ss.iloc[:,0],ss.iloc[:,1], color = 'green')
axes[1].set_xlabel(ss.columns[0])
axes[1].set_ylabel(ss.columns[1])

axes[2].bar(le.iloc[:,0],le.iloc[:,1], color = 'green')
axes[2].set_xlabel(le.columns[0])
axes[2].set_ylabel(le.columns[1])

fig.autofmt_xdate(rotation = 90)
plt.tight_layout()
```
